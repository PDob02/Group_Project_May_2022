```python
# Initial imports
import pandas as pd
from config import db_password
from sqlalchemy import create_engine
```

```python
# Import file
# https://www.kaggle.com/datasets/fmejia21/demographics-of-academy-awards-oscars-winners
file_path = 'Oscars_demographics.csv'
demo_df = pd.read_csv(file_path, encoding="ISO-8859-1")
```

```python
demo_df.head()
```

```python
# Clean demographic dataset
demo_df = demo_df.drop(columns=["_unit_id", "_golden", "_unit_state", "_trusted_judgments", "_last_judgment_at", "race_ethnicity:confidence", "religion", "religion:confidence", "sexual_orientation", "sexual_orientation:confidence", "year_of_award", "year_of_award:confidence", "birthplace:confidence", "date_of_birth:confidence", "award", "race_ethnicity", "biourl", "birthplace_gold", "date_of_birth_gold", "movie", "race_ethnicity_gold", "religion_gold", "religion_gold", "sexual_orientation_gold", "year_of_award_gold"], axis=1)
```

```python
# Rearrange column order
demo_df = demo_df[["person", "birthplace", "date_of_birth"]]
```

```python
demo_df['date_of_birth'] = pd.to_datetime(demo_df['date_of_birth'], errors='coerce')
```

```python
# Connection string to PostgreSQL
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/group_project"
```

```python
# Create database engine
engine = create_engine(db_string)
```

```python
# Export actor demoraphics (demo_df) to a SQL table
demo_df.to_sql(name="actor_demo", con=engine, index=False)
```
